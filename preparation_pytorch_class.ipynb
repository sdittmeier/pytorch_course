```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random x values
x = np.random.rand(100)

# Calculate y = x^2
y = x

# Print the dataset
print("x values:", x)
# Add random noise to y
noise = np.random.normal(0, 0.1, len(y))
y_with_noise = y + noise

# Print the dataset with noise
print("x values:", x)
print("y values with noise:", y_with_noise)
# Set the random seed for reproducibility
```

```python
# Plot the dataset
plt.scatter(x, y_with_noise)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title('Regression task')
plt.show()
```

```python
# Fit a 2nd order polynomial
coefficients = np.polyfit(x, y_with_noise, 1)

x_fit = np.linspace(0, 1, 100)
# Evaluate the polynomial at each data point
y_fit = np.polyval(coefficients, x_fit)

# Plot the dataset with the linearly regressed fit
plt.scatter(x, y_with_noise)
plt.plot(x_fit, y_fit, color='red', label='Linearly Regressed Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title('Regression task')
plt.legend()
plt.show()
```
